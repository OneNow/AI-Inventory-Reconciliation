```python
import torch
import matplotlib.pyplot as plt
import numpy as np

path = 'snapshots/resnet34_11.pth.tar'
checkpoint = torch.load(path)

train_loss_list = checkpoint['train_loss_list']
val_acc_list = checkpoint['val_acc_list']
train_acc_list = checkpoint['train_acc_list']
#val_loss_list = checkpoint['val_loss_list']

x = np.asarray(range(len(train_loss_list)))+1
x = x/2000.0
plt.plot(x[10:],train_loss_list[10:], color='green')
plt.xlabel('Number of Epochs')
plt.ylabel('Log Loss')
plt.title('Training Loss')
plt.show()

# print(val_loss_list[-1])
# x = np.asarray(range(len(val_loss_list)))+1
# x = x/2000.0
# plt.plot(x[10:],val_loss_list[10:], color='red')
# plt.xlabel('Number of Epochs')
# plt.ylabel('Log Loss')
# plt.title('Validation Loss')
# plt.show()

x = np.asarray(range(20))+1
plt.figure(1)
plt.plot(x, train_acc_list, 'green')
plt.xlabel('Number of Epochs')
plt.ylabel('Train Accuracy')
plt.title('Training accuracy curve')
plt.legend('Training')
#plt.show()

x = np.asarray(range(20))+1
plt.figure(1)
plt.plot(x, val_acc_list, 'red')
plt.xlabel('Number of Epochs')
plt.ylabel('Train/Val Accuracy')
plt.title('Train/Validation accuracy curve')
plt.legend('TV')
plt.show()
```
